```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("datasets\coin_Bitcoin.csv")
```

```python
df.head()
```

What is the average closing price of Bitcoin over the given dates?

```python
mean_close_price = df['Close'].mean()
print("average closing price: ", mean_close_price)
```

Visualization of daily closing price

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date_only'] = df['Date'].dt.date
df['Year'] = df['Date'].dt.year

# Select the first 20 days of data (adjust according to your actual dataset)
df_subset = df.head(800)

# Plot the line chart
plt.figure(figsize=(20,8))
sns.lineplot(data=df_subset, x='Date_only', y='Close', marker='o', color='red')

# Customize the plot
plt.title('Daily Closing Price of Bitcoin (First 20 Days)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid for easier visualization

# Show the plot
plt.tight_layout()
plt.show()
```
